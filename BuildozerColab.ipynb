## ④ Copiar el proyecto desde Drive
**Antes de ejecutar:** subí la carpeta `economia_hogar` (o el zip) a tu Google Drive.

Ajustá `RUTA_PROYECTO` si la carpeta está en otro lugar.

```
import os, shutil

# ── Ajustá esta ruta si tu carpeta está en otro lugar ──
RUTA_PROYECTO = '/content/drive/MyDrive/economia_hogar'
DESTINO       = '/content/economia_hogar'

if os.path.exists(DESTINO):
    shutil.rmtree(DESTINO)

shutil.copytree(RUTA_PROYECTO, DESTINO)
os.chdir(DESTINO)
print('✅ Proyecto copiado a:', DESTINO)
print('Archivos encontrados:')
for f in os.listdir(DESTINO):
    print(' ', f)
```

## ⑤ Crear buildozer.spec

```
spec = '''
[app]
title = Economia del Hogar
package.name = economiadelhogar
package.domain = org.miapp
source.dir = .
source.include_exts = py,png,jpg,kv,atlas,db
version = 1.0
requirements = python3,kivy==2.2.1,kivymd==1.1.1,matplotlib
orientation = portrait
fullscreen = 0
android.permissions = WRITE_EXTERNAL_STORAGE,READ_EXTERNAL_STORAGE
android.api = 33
android.minapi = 21
android.ndk = 25b
android.arch = arm64-v8a
android.allow_backup = True
p4a.branch = master

[buildozer]
log_level = 2
warn_on_root = 1
'''.strip()

with open('buildozer.spec', 'w') as f:
    f.write(spec)

print('✅ buildozer.spec creado')
print(open('buildozer.spec').read())
```

## ⑥ Compilar el APK ⏳
**Esta celda tarda entre 20 y 40 minutos la primera vez.**
Las siguientes compilaciones son más rápidas (~5 min) porque el SDK queda cacheado.

Si ves `BUILD SUCCESSFUL` al final, ¡todo salió bien!

```
!buildozer android debug 2>&1
```

## ⑦ Verificar y descargar el APK

```
import os, glob
from google.colab import files

apks = glob.glob('/content/economia_hogar/bin/*.apk')

if apks:
    apk = apks[0]
    size_mb = os.path.getsize(apk) / (1024 * 1024)
    print(f'✅ APK generado: {apk}')
    print(f'   Tamaño: {size_mb:.1f} MB')
    print('⬇️  Descargando...')
    files.download(apk)
else:
    print('❌ No se encontró el APK. Revisá los errores de la celda anterior.')
```

## ⑧ (Opcional) Guardar APK en Drive
Por si la descarga directa falla o querés conservarlo.

```
import shutil, glob

apks = glob.glob('/content/economia_hogar/bin/*.apk')
if apks:
    destino = '/content/drive/MyDrive/economiadelhogar.apk'
    shutil.copy(apks[0], destino)
    print(f'✅ APK guardado en Drive: {destino}')
else:
    print('❌ No se encontró el APK.')
```
